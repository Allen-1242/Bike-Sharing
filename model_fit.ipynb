```python
x=df_final.drop(['Date_start','Count'],axis=1)
y=df_final['Count']
x=x.astype('float')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn import svm
from math import sqrt
from sklearn.metrics import mean_squared_error
clf = svm.SVC()
print(clf.fit(x_train, y_train)) 
prediction1 = clf.predict(x_test)
prediction2 = clf.predict(x_train)
rms = sqrt(mean_squared_error(y_test, prediction1))
print("RMSE_test:",rms)
rms_1 = sqrt(mean_squared_error(y_train,prediction2))
print("RMSE_train:",rms_1)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
print(clf.fit(x_train,y_train))
prediction1 = clf.predict(x_test)
prediction2 = clf.predict(x_train)
rms = sqrt(mean_squared_error(y_test, prediction1))
print("RMSE_test:",rms)
rms_1 = sqrt(mean_squared_error(y_train,prediction2))
print("RMSE_train:",rms_1)
```

```python
from sklearn.cluster import KMeans
from matplotlib import style
style.use("ggplot")
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=10)
print(kmeans.fit(x_train,y_train))
prediction1 = clf.predict(x_test)
prediction2 = clf.predict(x_train)
rms = sqrt(mean_squared_error(y_test, prediction1))
print("RMSE_test:",rms)
rms_1 = sqrt(mean_squared_error(y_train,prediction2))
print("RMSE_train:",rms_1)
```

```python
from sklearn.neighbors.nearest_centroid import NearestCentroid
clf = NearestCentroid()
print(clf.fit(x_train, y_train)) 
prediction1 = clf.predict(x_test)
prediction2 = clf.predict(x_train)
rms = sqrt(mean_squared_error(y_test, prediction1))
print("RMSE_test:",rms)
rms_1 = sqrt(mean_squared_error(y_train,prediction2))
print("RMSE_train:",rms_1)
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier()
print(clf.fit(x_train, y_train))
prediction1 = clf.predict(x_test)
prediction2 = clf.predict(x_train)
rms = sqrt(mean_squared_error(y_test, prediction1))
print("RMSE_test:",rms)
rms_1 = sqrt(mean_squared_error(y_train,prediction2))
print("RMSE_train:",rms_1)
```
